# Loading Data

# > Column Descriptions:
    * year: The year the vehicle was manufactured.
    * make: The manufacturer of the vehicle (e.g., Kia, BMW, Volvo).
    * model: The specific model of the vehicle (e.g., Sorento, 3 Series, S60).
    * trim: The trim level of the vehicle, indicating the specific version or configuration (e.g., LX, 328i SULEV, T5).
    * body: The body type of the vehicle (e.g., SUV, Sedan).
    * transmission: The type of transmission in the vehicle (e.g., automatic).
    * vin: The Vehicle Identification Number, a unique code used to identify individual vehicles.
    * state: The state where the vehicle was registered or sold (e.g., CA).
    * condition: A numerical score indicating the condition of the vehicle, with higher values generally representing better condition.
    * odometer: The mileage of the vehicle, indicating how many miles it has been driven.
    * color: The exterior color of the vehicle.
    * interior: The interior color of the vehicle.
    * seller: The entity or individual selling the vehicle (e.g., "kia motors america inc").
    * mmr: The Manheim Market Report (MMR) value, which provides a wholesale market price estimate for the vehicle.
    * sellingprice: The actual price at which the vehicle was sold.
    * saledate: The date and time when the sale occurred.

# Data Preprocessing

```python
data.info()
```

```python
data['saledate']=pd.to_datetime(data['saledate'],errors='coerce')
```

```python
data.info()
```

```python
data['day']=data['saledate'].dt.day_name()
data
```

```python
data.isna().sum()
```

```python
data[data.duplicated()]
```

```python
# 1.distribution of transmission 
data.groupby(['transmission'])['condition'].mean().plot(kind='bar')
plt.title('how transmission affects the condition of vehicle');
```

* The type of the transmission does not effect condition.
* both automatic and manual vehicle has almost same condition 

```python
# 2.Most revenue generating product:
data.groupby(['model','make'])['sellingprice'].max().sort_values(ascending=False).head(30).plot(kind='bar',color='skyblue')
plt.title('Most Revenue Generating Model');
```

* The model Escape introduced by Ford was generating more revenue .

```python
# 3.Most commonly sold car companies
data['make'].value_counts().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Most commonly sold car companies');
```

* The ford has sold more products as compared to others

```python
# 4.top 20 high price body type
data.groupby(['body'])['sellingprice'].max().sort_values(ascending=False).head(20).plot(kind='bar',color='white',edgecolor='black')
plt.title('Top 20 High Revenue Generating Body Type')
plt.xlabel('Body Type')
plt.ylabel('Price');
```

* The body type SUV has highest price in the market


```python
# 5.Average revenue according to color .
data.groupby(['color'])['mmr'].mean().sort_values(ascending=False).head(20).plot(kind='bar',color='red',edgecolor='black')
plt.title('Average price of each body color in mmr ');
```

* the car with charcoal color has highest average revenue
* green and gold has least average revenue

```python
# 6.Most frequently sold model:
data['model'].value_counts().head(5).plot(kind='pie',autopct='%1.1f%%')
plt.title('Most Frequently Sold Model');
```

* Alitima was sold more frequently

```python
#7.Distribution of market price of car in each state:
data.groupby(['state','model'])['sellingprice'].sum().sort_values(ascending=False).head(20).plot(kind='bar',color='yellow',edgecolor='black')
plt.title('Top car model according to state');
```

* the Altima intrduced in Florida has highest price in market
